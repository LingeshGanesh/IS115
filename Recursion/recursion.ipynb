# **Recursion**

A recursion function calls itself to solve the smaller pieces before adding it up to get a final result. It is comprised of a base case and reductions steps.

- **Base Case:** The stopping point so that the function does not go infinite
- **Reduction Step:** The function to be performed on smaller instances of the same problem

### Example: Fibonacci's Sequence

Fibonacci(n):
    if n <= 1:  // Base case: Return n if it is 0 or 1
        return n
    return Fibonacci(n-1) + Fibonacci(n-2)  // Recursive case: Sum the two previous Fibonacci numbers

```python
def fibonacci(n):
    if n <= 1:  # Base case: Fibonacci of 0 or 1 is the number itself
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)  # Recursive case: sum of previous two Fibonacci numbers

# Example usage:
n = 5
print(f"Fibonacci of {n} is: {fibonacci(n)}")
```
