# **Linear Search**

Linear search is a simple searching algorithm that goes through each element of an array iteratively to see if it matches a target.

The time complexity of Linear Search has a best case of O(1) which would mean that the first element of the array is the target. It has a worst case time complexity of O(n) as the element would either not be in the array or would be the last element of the array.

## Flow

1) Initialize pointer at the first element of the array
2) Compare the element at the pointer with the target
    - If it matches, return the index
    - If it doesn't match, increase the pointer index by 1
3) Repeat step 2 until the target is found or the whole array has been traversed

## Implementation

```python
LinearSearch(A, target):
    for i = 0 to len(A) - 1:  // Iterate through each element in the array
        if A[i] == target:     // Check if the current element is the target
            return i            // If found, return the index of the target
    return -1                   // If the target is not found, return -1
```
