# Insertion Sort

https://www.geeksforgeeks.org/dsa/insertion-sort-algorithm/

Insertion sort is a sorting algorithm that works by iteratively inserting each element of an unsorted list into its correct position in a sorted position of the list.

Insertion sort has a best case time complexity of O(n), as even if the list has already been fully sorted, it would still have to go through all n items of an array. It has a worst case time complexity of O(n^2) due to the worst case scenario (the array is sorted in reverse order) requiring each element to be shifted to the beginning.

## Flow

1) Initialize pointer at the second element of the list (The first element is considered to be sorted by default)
2) Compare the element to the element to the left of it
    - If the current element is smaller, swap the places of the two
    - THE POINTER DOES NOT CHANGE
3) Repeat step 2 until the current element is greater than the one to its left (all other elements to its left should be sorted already)
4) Increase the pointer value by 1
5) Repeat steps 2-4 until pointer is PAST the last element of the array

## Implementation

```python
InsertionSort(A):
    for i = 1 to len(A) - 1:  // Start with the second element (index 1)
        key = A[i]             // The current element to be inserted
        j = i - 1              // Start comparing with the element before it

        // Shift elements of A[0...i-1] that are greater than key to the right
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]    // Shift element to the right
            j = j - 1

        // Insert the key at the correct position
        A[j + 1] = key
```
