# **Queues**

A queue is a linear data structure, meaning that only one data element is accessible at any point of time.

The stack has a **FIFO (First In, First Out) or FCFS (First Come, First Serve)** property, meaning that the *first* item placed on the stack would be the *first* item removed.

It has 3 main operations:

1) enqueue
2) dequeue
3) peek

```python
from collections import deque
```

### Enqueue
Places a new data element to the tail of the queue

```python
queue = deque()

queue.append(10)
queue.append(20)
queue.append(30)

print(queue)
```

### Dequeue

Removes the data element at the head of the queue

```python
x = queue.popleft()

print(x)
print(queue)
```

### Peek
Returns the first data element of the queue without removing it

```python
print(queue[0])
```
