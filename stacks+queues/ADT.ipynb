# **Abstract Data Types (ADT)**

An abstract data type is a data type who's representation is hidden from the client

- Defined by the operations that it supports
- NOT defined by specific implementations

This is the opposite of a Concrete Data Structure, which describes how the data is actually stored and managed, including implementation details.



## **Advantages**

### Encapsulation
The implementation may be complex, but clients do not need to know

### Localization
Any change to the implementation should not affect existing clients as only the result would be returned to them

### Flexibility
We can swap different implementations of the same ADT without affecting the client
